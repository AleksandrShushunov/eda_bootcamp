### Заполним пропуски в данных

Найдем среднее время, проводимое пользователями на одной странице типов Informational и ProductRelated и умножим полученное значение на число посещенных страниц для нахождения суммарного времени, проведенного на страницам этих типов

```python
inf_dur = df.Informational_Duration.mean() / df.Informational.mean() 
```

```python
df['Informational_Duration'] = df['Informational_Duration'].fillna(df.Informational * inf_dur)
```

```python
prod_dur = df.ProductRelated_Duration.mean() / df.ProductRelated.mean()
```

```python
df['ProductRelated_Duration'] = df['ProductRelated_Duration'].fillna(df.ProductRelated * prod_dur)
```

Признак ExitRates имеет крайне высокую корреляцию(0.91) с признаком BounceRates, заполнять пропуски в данном столбце не будем, удалим его для улучшения качества модели

```python
df.drop('ExitRates',axis = 1, inplace = True)
```
